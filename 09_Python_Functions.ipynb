```python
#global variable
#declare out side any function and accessible throghout the program
counter = 0
def increment():
  global counter
  counter += 1
  print(f"COUNTER INCREMENT TO {counter}")
```

```python
def reset():
  global counter
  counter = 0
  print(f"COUNTER RESET TO {counter}")
```

```python
#local variable
#declare inside a function and accessible only with that function
def increment():
  counter = 0
  counter += 1
  print(f"COUNTER INCREMENT TO {counter}")
```

```python
#round key word
def principal_amount (p,r,n,t):
  A = p * (1 + (r/n))**(n-t)
  return A
round(principal_amount(1000,0.05,12,5),2)
```

```python
def EMICALL(p,rR,nN):
  monthly_rate = (rR/100)/12
  total_amount = nN * 12
  emi = (p*monthly_rate*((1+monthly_rate)**total_amount))/(((1+monthly_rate)**total_amount)-1)
  return emi
round(EMICALL(500000,7.5,20))
```
