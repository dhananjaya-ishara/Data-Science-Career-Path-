#### Problem 1:

You are given a JSON file named students.json that contains information about students in a class. Each student has attributes such as name, age, grades, and is_enrolled.

Your task is to:
1. Read the JSON file and load its data into a Python dictionary.
2. Print the name and grades of all students.

```python
import json

# Step 1: Read the JSON file and load it into a Python list (of dictionaries)
# Remember to replace 'students.json' with the actual path to your JSON file
with open('students.json', 'r') as file:
    students_data = json.load(file)

# Step 2: Print the name and grades of all students
for student in students_data:
    name = student.get('name')
    grades = student.get('grades')
    print(f"Name: {name}, Grades: {grades}")
```

#### Problem 2:

You are given an XML file named employees.xml that contains information about employees in a company. Each employee has attributes such as name, position, salary, and is_manager.

Your task is to:
1. Parse the XML file and retrieve the root element.
2. Print the names and positions of all employees.

```python
import xml.etree.ElementTree as ET

#reading xml data from a file
tree = ET.parse('employees.xml')#parsing the xml file into a tree structure
root = tree.getroot()

#accessing the text value of a specific tag
print(root.find('name').text)
print(root.find(' position').text)
print(root.find('salary').text)
print(root.find('manager').text)
```
